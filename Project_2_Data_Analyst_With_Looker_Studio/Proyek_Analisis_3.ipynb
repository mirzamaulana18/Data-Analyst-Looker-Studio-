```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv("train.csv")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
print("jumlah duplikasi : ", df.duplicated().sum())
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
datetime_columns = ["Order Date", "Ship Date"]

for column in datetime_columns:
    df[column] = pd.to_datetime(df[column], format='%d/%m/%Y')
```

```python
df.info()
```

```python
#Pisahkan tahu, bulan dan hari dari Order Date dan Ship Date untuk kemudahan analisis
df['Order Date'] = pd.to_datetime(df['Order Date'])

# Tambah kolom baru Year, Month, Day
df['Year'] = df['Order Date'].dt.year
df['Month'] = df['Order Date'].dt.month
df['Day'] = df['Order Date'].dt.day

df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Tambah kolom baru Year, Month, Day
df['Year'] = df['Ship Date'].dt.year
df['Month'] = df['Ship Date'].dt.month
df['Day'] = df['Ship Date'].dt.day
```

```python
df.head()
```

```python
df.drop(columns=['Row ID', 'Customer Name', 'Postal Code'], inplace=True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.to_csv('train_clean.csv', index=False)
```
