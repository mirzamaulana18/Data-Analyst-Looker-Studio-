```python
#Bedah kolom TotalCharges karena menurut info dataset, ada nilai / spasi koson yan terhitung sebagai nilai
# lihat beberapa contoh isi
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

# cek baris yang gagal konversi
print(df[df['TotalCharges'].isna()])

```

```python
#Ubah kolom ke numeric, nilai kosong/spasi otomatis jadi NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
#Cek berapa banyak nilai NaN (kosong setelah dibersihkan)
print("Jumlah missing:", df['TotalCharges'].isna().sum())
print(df[df['TotalCharges'].isna()][['TotalCharges']].head())
```

```python
#(Opsional) Isi nilai kosong dengan median/mean atau drop
# contoh: isi dengan median
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

```python
print("Jumlah NaN:", df['TotalCharges'].isna().sum())
print(df[df['TotalCharges'].isna()][['TotalCharges']])
```

```python
if df['TotalCharges'].dtype == 'object':
    mask_spasi = df['TotalCharges'].str.strip() == ''
    print("Jumlah spasi kosong:", mask_spasi.sum())
    print(df[mask_spasi][['TotalCharges']])
else:
    print("Kolom TotalCharges sudah numerik, jadi tidak mungkin ada spasi kosong.")
```
